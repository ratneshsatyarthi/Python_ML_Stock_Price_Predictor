```python
# Saving the Extracted Data of TCS in Variable
```

```python
Adj_Close_Price = TCS_data[['Adj Close']]
```

```python
# Finding Max and Min values
```

```python
max(Adj_Close_Price.values),min(Adj_Close_Price.values)
```

```python
# Normalizing the Data between 0 to 1
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Adj_Close_Price)
scaled_data
```

```python
# View Length of the scaled data - should be same as present in starting
```

```python
len(scaled_data)
```

```python
# Taking Input Trainnig data and Predicting next Data
```

```python
x_data = []
y_data = []

for i in range (100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
# X data has 100 values and Y values has 1 predicted value
```

```python
x_data[0], y_data[0]
```

```python
# Split data in to Training(70%) and Testing(30%) Dataset
```

```python
int(len(x_data)*0.7)
```

```python
2463-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
# Print the Training and Test Dataset
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Build Neural Network Model to predict the price
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences = True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(64, return_sequences = False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Train Dataset into multipple batch and Epoch
```

```python
model.fit(x_train, y_train, batch_size=1, epochs = 2)
```

```python
# View Summary
```

```python
model.summary()
```

```python
# Predicting the price based on test data
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
# Invert the price - Derationalizing
```

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
# Check Root Mean Square Value for Accuarcy
```

```python
rmse = np.sqrt(np.mean (( inv_predictions - inv_y_test)**2))
```

```python
rmse
```

```python
# Plot Original values vs Predicted values
```

```python
import pandas as pd
plotting_data = pd.DataFrame(
    {
        'original_test_data': inv_y_test.reshape(-1),
        'predictions': inv_predictions.reshape(-1)
    },
    index = TCS_data.index[splitting_len+100:]
)
plotting_data.head()
```
